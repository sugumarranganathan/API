```python
from google.colab import userdata

GEMINI_API_KEY = userdata.get("GEMINI_API_KEY")

print("API Key Loaded Successfully!")
```

Create Client

```python
from google import genai

client = genai.Client(api_key=GEMINI_API_KEY)
```

Interactive Chatbot

```python
# Ask what type of chatbot to create
system_msg = input("What type of chatbot would you like to create?\n\n")

print("\nYour new assistant is ready!")
print("Type 'quit' to exit.\n")

chosen_model = "models/gemini-3.5-flash"
# or:
# chosen_model = "gemini-flash-latest"

while True:

    user_input = input("You : ")

    if user_input.lower() == "quit":
        print("Goodbye!")
        break

    prompt = f"""
You are a {system_msg}.

User:
{user_input}
"""

    response = client.models.generate_content(
        model=chosen_model,
        contents=prompt
    )

    print("\nGemini :", response.text)
    print("-" * 70)
```
