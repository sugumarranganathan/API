```python
from openai import OpenAI
from google.colab import userdata

# Read Groq API Key from Colab Secret
api_key = userdata.get("GROQ_API_KEY")

# Initialize Groq Client
client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

def get_completion(prompt, model="llama-3.3-70b-versatile"):

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )

    return response.choices[0].message.content
```

The second code defines a reusable function that sends a user prompt to the AI model, receives the generated response, and returns it to the application.

Helper function (chat format)

Here's the helper function we'll use in this course.

===

This function sends an entire conversation (system, user, and assistant messages) to the AI model, generates a context-aware response, and returns the AI's reply.

It is mainly used to build chatbots and conversational AI applications that remember previous messages and provide natural multi-turn conversations.

```python
from openai import OpenAI
from google.colab import userdata

# Read Groq API Key from Colab Secret
api_key = userdata.get("GROQ_API_KEY")

# Initialize Groq Client
client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

def get_completion_from_messages(
    messages,
    model="llama-3.3-70b-versatile",
    temperature=0,
    max_tokens=500
):

    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,      # Controls randomness
        max_tokens=max_tokens         # Maximum output tokens
    )

    return response.choices[0].message.content
```

This helper function sends a conversation to the Groq AI model, generates a response, and also returns the number of tokens used for the input, output, and total request.

```python
from openai import OpenAI
from google.colab import userdata

# Read Groq API Key from Colab Secret
api_key = userdata.get("GROQ_API_KEY")

# Initialize Groq Client
client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

def get_completion_and_token_count(
    messages,
    model="llama-3.3-70b-versatile",
    temperature=0,
    max_tokens=500
):

    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )

    # AI Generated Response
    content = response.choices[0].message.content

    # Token Usage
    token_dict = {
        "prompt_tokens": response.usage.prompt_tokens,
        "completion_tokens": response.usage.completion_tokens,
        "total_tokens": response.usage.total_tokens,
    }

    return content, token_dict
```

| **Step** | **Process**       | **Description**                                              |
| -------- | ----------------- | ------------------------------------------------------------ |
| **1**    | Read API Key      | Retrieves the Groq API key from Colab Secrets.               |
| **2**    | Create Client     | Initializes the Groq API client.                             |

| **4**    | Send Request      | Sends the conversation to the Groq model.                    |
| **5**    | Generate Response | The AI processes the request and creates a response.         |
| **6**    | Count Tokens      | Records the number of input, output, and total tokens.       |
| **7**    | Return Results    | Returns both the AI response and the token usage dictionary. |

```python
messages = [
    {'role': 'system', 'content': "You are an assistant who responds in the style of Dr Seuss."},
    {'role': 'user', 'content': "Write me a very short poem about a happy carrot"}
]

response, token_dict = get_completion_and_token_count(messages)

print("Response:", response)
print("Token Usage:", token_dict)
```

This code sends a conversation (system and user messages) to the AI model, receives the generated response, and also retrieves the token usage statistics for the request.

It demonstrates how to generate an AI response while monitoring the number of input tokens, output tokens, and total tokens consumed by the API call.
