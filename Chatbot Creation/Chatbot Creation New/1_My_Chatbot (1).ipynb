```python
from google.colab import userdata
from openai import OpenAI

client = OpenAI(
    api_key=userdata.get("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "user", "content": "Hello!"}
    ]
)

print(response.choices[0].message.content)
```

Many beginners think the AI should simply repeat "Hello!", but that's not how Large Language Models (LLMs) work.

What is messages?

messages is the conversation that you send to the AI.

Think of it as a chat history.

For example:

You  → Hello!
AI   → Hello. It's nice to meet you...

You only send your side of the conversation.

The AI generates the next reply.

What is role?

The role tells the AI who is speaking.

There are three main roles.

====

| Role        | Who is speaking? | Purpose                      |
| ----------- | ---------------- | ---------------------------- |
| `system`    | The developer    | Gives instructions to the AI |
| `user`      | You              | Your question or request     |
| `assistant` | The AI           | The AI's previous response   |

An LLM is not a lookup table that returns a fixed answer. It generates a response based on the conversation you provide. When you send "Hello!", you're not asking it to repeat the word—you are giving it the first message in a conversation, and the model generates the most appropriate next message. That's why a short input like "Hello!" can naturally produce a longer, friendly response.



```python
from google.colab import userdata
from openai import OpenAI

client = OpenAI(
    api_key=userdata.get("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "user", "content": "Tell me a joke."}
    ]
)

print(response.choices[0].message.content)
```

```python
from openai import OpenAI
from google.colab import userdata

# Read API key from Colab Secret
api_key = userdata.get("GROQ_API_KEY")

# Create Groq client
client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

messages = []

system_msg = input("What type of chatbot would you like to create?\n")

messages.append({
    "role": "system",
    "content": system_msg
})

print("\nYour new assistant is ready!")
print("Type 'quit' to exit.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "quit":
        print("Goodbye!")
        break

    messages.append({
        "role": "user",
        "content": user_input
    })

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        temperature=0.7
    )

    reply = response.choices[0].message.content

    messages.append({
        "role": "assistant",
        "content": reply
    })

    print(f"\nAssistant: {reply}\n")
```

```python
from openai import OpenAI
from google.colab import userdata
import gradio as gr

# Read Groq API Key from Colab Secret
api_key = userdata.get("GROQ_API_KEY")

# Initialize Groq Client
client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

# Conversation Memory
messages = [
    {
        "role": "system",
        "content": "You are a financial expert specializing in real estate investment and negotiation."
    }
]

# Chat Function
def CustomChatGPT(user_input):

    messages.append({
        "role": "user",
        "content": user_input
    })

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        temperature=0.7,
        max_tokens=1024
    )

    reply = response.choices[0].message.content

    messages.append({
        "role": "assistant",
        "content": reply
    })

    return reply

# Gradio Interface
demo = gr.Interface(
    fn=CustomChatGPT,
    inputs=gr.Textbox(
        lines=4,
        placeholder="Ask your real estate investment question..."
    ),
    outputs=gr.Textbox(lines=8),
    title="🏠 My Saving Assistant",
    description="AI Financial Expert powered by Groq + Llama 3.3 70B"
)

# Launch
demo.launch(share=True)
```
