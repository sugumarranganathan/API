"L2: Evaluate Inputs: Classification" is about checking and classifying a user's input before sending it to the AI model. This is an important concept in AI applications because not every user message should be handled the same way.

What is "Evaluate Inputs"?

Evaluate Inputs means:

Analyze the user's message first to understand what type of request it is.

Instead of immediately asking GPT to answer, your application first asks:

What is the user trying to do?
Is it a question?
Is it a complaint?
Is it asking for a refund?
Is it asking for technical help?
Is it inappropriate?
Should this message go to another system?

This step is called input evaluation.

User Message
      │
      ▼
Evaluate Input
      │
      ▼
Classify
      │
      ▼
Choose the correct workflow
      │
      ▼
Call GPT or another API
      │
      ▼
Final Response

his technique is widely used in production AI systems, including:

Customer Support Chatbots – Route users to sales, billing, or technical support.
Email Classification – Label emails as spam, promotions, or important.
Help Desk Ticketing – Automatically assign support tickets to the right team.
Banking Apps – Detect fraud reports, loan requests, or account inquiries.
Healthcare Systems – Categorize appointment requests, prescription renewals, or emergency messages.
AI Agents – Decide which tool or API (search, calculator, database, etc.) should be used to fulfill the user's request.

In short, Evaluate Inputs is the process of understanding what the user wants, and Classification is the process of assigning that request to a predefined category so your application can respond in the most appropriate way.

Function

```python
def get_completion_from_messages(
    messages,
    model="llama-3.3-70b-versatile",
    temperature=0,
    max_tokens=500
):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )

    return response.choices[0].message.content
```

```python
delimiter = "####"

system_message = f"""
You will be provided with customer service queries.
The customer service query will be delimited with
{delimiter} characters.

Classify each query into a primary category
and a secondary category.

Provide your output in json format with the
keys: primary and secondary.

Primary categories:
Billing, Technical Support,
Account Management,
or General Inquiry.

Billing secondary categories:
Unsubscribe or upgrade
Add a payment method
Explanation for charge
Dispute a charge

Technical Support secondary categories:
General troubleshooting
Device compatibility
Software updates

Account Management secondary categories:
Password reset
Update personal information
Close account
Account security

General Inquiry secondary categories:
Product information
Pricing
Feedback
Speak to a human
"""

user_message = "I want you to delete my profile and all of my user data"

messages = [
    {
        "role": "system",
        "content": system_message
    },
    {
        "role": "user",
        "content": f"{delimiter}{user_message}{delimiter}"
    }
]

response = get_completion_from_messages(messages)
print(response)
```

| Code       | Purpose                                   | Why it is needed                                                                                                               |
| ---------- | ----------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
| **Code 1** | Initialize the Groq client                | Connects your Colab notebook to the Groq API using your API key. Without this, your notebook cannot communicate with the LLM.  |
| **Code 2** | `get_completion_from_messages()` function | Sends your prompt (`messages`) to the model and returns the generated response. This avoids repeating the API call every time. |


```python
user_message = f"""\
Tell me more about your flat screen tvs"""
messages =  [
{'role':'system',
 'content': system_message},
{'role':'user',
 'content': f"{delimiter}{user_message}{delimiter}"},
]
response = get_completion_from_messages(messages)
print(response)
```

| **Code**                                                              | **Purpose**                                                                                                  | **Input**                     | **Output**                                                              |
| --------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------ | ----------------------------- | ----------------------------------------------------------------------- |
| `user_message = f"""\ Tell me more about your flat screen tvs"""`     | Stores the customer's query in a variable.                                                                   | Customer's question           | `user_message` contains `"Tell me more about your flat screen tvs"`     |
| `messages = [`                                                        | Creates a conversation list that will be sent to the LLM.                                                    | System message + User message | A list containing two dictionaries (`system` and `user`)                |
| `{'role':'system', 'content': system_message}`                        | Provides instructions to the AI about how it should behave and classify the query.                           | `system_message`              | AI understands the classification rules.                                |
| `{'role':'user', 'content': f"{delimiter}{user_message}{delimiter}"}` | Sends the customer's query to the AI. The delimiter (`####`) separates the user input from the instructions. | `user_message`                | User query wrapped as `####Tell me more about your flat screen tvs####` |
| `response = get_completion_from_messages(messages)`                   | Sends the complete conversation to the Groq/OpenAI model and waits for a response.                           | `messages` list               | AI-generated response                                                   |
| `print(response)`                                                     | Displays the AI's response on the screen.                                                                    | `response`                    | Prints the JSON classification                                          |
